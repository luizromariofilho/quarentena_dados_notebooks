```python
import pandas as pd
```

```python
movies = pd.read_csv("./dados/movies.csv")
```

```python
movies.head()
```

```python
ratings = pd.read_csv("dados/ratings.csv")
```

```python
ratings.head()
```

```python
means_ratings_by_movies = ratings.groupby("movieId")["rating"].mean()
```

```python
means_ratings_by_movies.head()
```

```python
movies_with_mean = movies.join(means_ratings_by_movies, on="movieId")
movies_with_mean.head()
```

```python
movies_with_mean.sort_values("rating", ascending=False)
```
